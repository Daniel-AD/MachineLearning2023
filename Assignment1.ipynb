## Feed Forward

```python
def feed_forward(a0,w,b):
    a = a0
    num_layers = len(b)
    
    # feed input layer forward
    for ℓ in range(num_layers):
        z = w[ℓ] @ a + b[ℓ]
        a = 1.0/(1.0 + np.exp(-z))
    return a
```

```python
#restructure and switch dimensions 
def feed_forward2(a0,w,b):
    num_layers = len(w)
    a = a0
    for ℓ in range(num_layers):
        z = np.dot(a,w[ℓ]) + b[ℓ]
        a = 1.0/(1.0+np.exp(-z))
    return a
```
